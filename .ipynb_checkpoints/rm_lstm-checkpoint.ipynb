```python
import numpy as np

with open('/root/rm-RNN/data/rm_trans.txt') as f:
    content = f.read()
    
output = []

for item in content:
    output.append(item)
    
```

```python
output
```

```python
# create dictionary for character vocab
chars = sorted(list(set(output)))
char_dict = dict((c,i) for i,c in enumerate(chars))
```

```python
n_chars = len(output)
n_vocab = len(char_dict)

print("Total # of Characters: {}".format(n_chars))
print("Total # of Vocab: {}".format(n_vocab))
```

```python
# prepare dataset for input

seq_length = 100
d_X = []
d_Y = []

for i in range(0, n_chars - seq_length, 1):
    seq_in = output[i:i + seq_length]
    seq_out = output[i + seq_length]
    d_X.append([char_dict[char] for char in seq_in])
    d_Y.append([char_dict[seq_out]])

n_patterns = len(d_X)

print("Total Patterns: {}".format(n_patterns))
```

```python
from keras.utils import np_utils

# reshape input -> [samples, time_steps, features]
input_X = np.reshape(d_X, (n_patterns, seq_length, 1))
# normalize data
input_X = input_X/float(n_vocab)

# one hot encode output
output_Y = np_utils.to_categorical(d_Y)
```

```python
# keras RNN model (LSTM)
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint

model = Sequential()

model.add(LSTM(256, input_shape=(input_X.shape[1], input_X.shape[2], return_sequences=True)))
model.add(Dropout(0.2))
model.add(LSTM(256))
model.add(Dense(output_Y.shape[1], activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
# training the model
checkpoint = ModelCheckpoint('/saved_models/weights_{epoch:02d}-{loss:.4f}.hdf5', monitor='loss', verbose=1, save_best_only=True)

model.fit(input_X, output_Y, epochs=25, batch_size=128, callbacks=[checkpoint]
```
